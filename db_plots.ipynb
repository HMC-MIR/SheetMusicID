```python
import seaborn as sns
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("/data1/kji/plot_data/fp_distribution_v0.3.1.pkl", "rb") as f:
    distribution_a = pickle.load(f)
```

```python
with open("/data1/kji/plot_data/fp_distribution_v0.3.1b.pkl", "rb") as f:
    distribution_b = pickle.load(f)
```

```python
with open("/data1/kji/plot_data/fp_distribution_v0.3.1c.pkl", "rb") as f:
    distribution_c = pickle.load(f)
```

```python
x_a = np.arange(len(distribution_a))
x_b = np.arange(len(distribution_b))
x_c = np.arange(len(distribution_c))
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1, 1, 1)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_title('Frequency Distribution of Fingerprints', fontsize=20)
ax.set_xlabel('Fingerprint Count', fontsize=15)
ax.set_ylabel('Log Frequency', fontsize=15)
ax.plot(x_a, distribution_a, label = "version 0.3.1")
ax.plot(x_b, distribution_b, label = "version 0.3.1b")
ax.plot(x_c, distribution_c, label = "version 0.3.1c")
plt.legend(loc="upper right",fontsize = 13)
fig.savefig("fp_distribution.png", dpi = 300)
```
